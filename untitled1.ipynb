## Lecture des données

```python
dataset_path = "/content/gdrive/My Drive/DATA_VE_Nour el Hamza.xlsx"
dataframe = pd.read_excel(dataset_path, sheet_name='détail')
valid_data = dataframe[dataframe.compliance=='valid']
valid_data.head()
```

## 1. Heure de début et fin de recharge par jour, mois et année : 
Analysez la répartition des heures de début et de fin de recharge par jour, mois et année. Vous pouvez créer des graphiques ou des histogrammes pour visualiser les tendances et les moments de pointe de la recharge.

```python
df = pd.DataFrame()
df['jour début charge'] = valid_data['date début de charge'].dt.day_name()
df['mois début charge'] = valid_data['date début de charge'].dt.month_name()
df['année début charge'] = valid_data['date début de charge'].dt.year
df['heure début charge'] = valid_data['date début de charge'].dt.hour
df['jour fin charge'] = valid_data['date fin de charge'].dt.day_name()
df['mois fin charge'] = valid_data['date fin de charge'].dt.month_name()
df['année fin charge'] = valid_data['date fin de charge'].dt.year
df['heure fin charge'] = valid_data['date fin de charge'].dt.hour
```

```python
df
```

### 1.1 Analyse par heure d'arrivée

#### Heure de début par jour

```python
for jour in jours_semaine:
    subset = df[df['jour début charge'] == jour]
    plt.hist(subset['heure début charge'], bins=range(24), alpha=1, label=jour)
plt.xlabel('heure début charge')
heures_arrivee = df['heure début charge'].dropna().unique()
plt.xticks(heures_arrivee)
plt.ylabel('Nombre de occurrences')
plt.title('Histogramme des heures d\'arrivée par jour de la semaine')
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(len(jours_semaine), 1, figsize=(10, 16), sharex=True)
for i, jour in enumerate(jours_semaine):
    subset = df[df['jour début charge'] == jour]
    axs[i].hist(subset['heure début charge'], bins=range(24), alpha=0.5)
    axs[i].set_title(jour)
fig.suptitle('Histogramme des heures d\'arrivée pour chaque jour de la semaine')
heures_arrivee = df['heure début charge'].dropna().unique()
plt.xticks(heures_arrivee)
plt.xlabel('Heure d\'arrivée')
plt.ylabel('Nombre d\'occurrences')
plt.tight_layout()
plt.show()
```

#### Heure d'arrivé par mois

```python
for mois in mois_annee:
    subset = df[df['mois début charge'] == mois]
    plt.hist(subset['heure début charge'], bins=range(24), alpha=1, label=mois)
plt.xlabel('heure début charge')
heures_arrivee = df['heure début charge'].dropna().unique()
plt.xticks(heures_arrivee)
plt.ylabel('Nombre de occurrences')
plt.title('Histogramme des heures d\'arrivée par mois de l\'année')
plt.legend()
plt.show()
```

#### Heure d'arrivée par année

```python
plt.hist(df['heure début charge'], bins=range(24), alpha=1)
plt.xlabel('heure début charge')
heures_arrivee = df['heure début charge'].dropna().unique()
plt.xticks(heures_arrivee)
plt.ylabel('Nombre de occurrences')
plt.title('Histogramme des heures d\'arrivée par annee (2019)')
plt.show()
```

## 2. Identifier les consommations:
 VE qui consomment le plus ou le moins, les périodes de l'année où la consommation est la plus élevée ou la plus faible, les facteurs qui influencent la durée de transaction ou de charge, etc.

```python
df = pd.DataFrame()
df['ves'] = valid_data['idTag']
df['cons'] = valid_data['consommation (kWh)']
df['period'] = valid_data['durée de charge']
df['jour'] = valid_data['date début de charge'].dt.day_name()
df['mois'] = valid_data['date début de charge'].dt.month_name()
```

### Consommation par Ve

#### consommation total par véhicule


```python
somme_consommation = df.groupby('ves')['cons'].sum()
plt.bar(somme_consommation.index, somme_consommation)
plt.ylabel('Consommation (Kwh)')
plt.title('Consommation des VEs pour l\'année 2019')
plt.show()
```

#### Consommation moyenne par véhicule

```python
somme_consommation = df.groupby('ves')['cons'].sum()
period_consommation = df.groupby('ves')['period'].sum()
cons = somme_consommation / period_consommation
plt.bar(cons.index, cons)
plt.ylabel('Consommation moyenne (KWH/min)')
plt.title('Consommation moyenne des VEs pour l\'année 2019')
plt.show()
```

### Consommation par période

#### Consommation total par jour

```python
somme_consommation = df.groupby('jour')['cons'].sum()
plt.bar(somme_consommation.index, somme_consommation)
plt.ylabel('Consommation (Kwh)')
plt.title('Consommation par jour de la semaine pour l\'année 2019')
plt.show()
```

#### Consommation total par mois

```python
somme_consommation = df.groupby('mois')['cons'].sum()
plt.bar(somme_consommation.index, somme_consommation)
plt.ylabel('Consommation (Kwh)')
plt.title('Consommation par mois pour l\'année 2019')
plt.show()
```

### Consommation par VE Par Période

#### Consommation par VE par jour



```python
consommation_par_jour = df.groupby(['jour', 'ves'])['cons'].sum().unstack()
for vehicule in consommation_par_jour.columns:
    plt.bar(consommation_par_jour.index, consommation_par_jour[vehicule], label=vehicule)
plt.ylabel('Consommation total (KWH)')
plt.title('Consommation des VEs par jour pour l\'année 2019')
plt.legend()
plt.show()
```

```python
consommation_par_ve = df.groupby(['ves', 'jour'])['cons'].sum().unstack()
for jour in consommation_par_ve.columns:
    plt.bar(consommation_par_ve.index, consommation_par_ve[jour], label=jour)
plt.ylabel('Consommation total (KWH)')
plt.title('Consommation journalière par VE pour l\'année 2019')
plt.legend()
plt.show()
```

#### Consommation par mois

```python
consommation_par_mois = df.groupby(['mois', 'ves'])['cons'].sum().unstack()
for vehicule in consommation_par_mois.columns:
    plt.bar(consommation_par_mois.index, consommation_par_mois[vehicule], label=vehicule)
plt.ylabel('Consommation total (KWH)')
plt.title('Consommation des VEs par mois pour l\'année 2019')
plt.legend()
plt.show()
```

```python
consommation_par_ve = df.groupby(['ves', 'mois'])['cons'].sum().unstack()
for mois in consommation_par_ve.columns:
    plt.bar(consommation_par_ve.index, consommation_par_ve[mois], label=mois)
plt.ylabel('Consommation total (KWH)')
plt.title('Consommation par mois pour chaque VE pour l\'année 2019')
plt.legend()
plt.show()
```

## 4.	Le taux d'occupation des bornes:
 de recharge par jour, par mois et par année. Utiliser la durée de transaction et le nombre de transactions pour chaque borne.

```python
df = pd.DataFrame()
df['period'] = valid_data['durée de charge']
df['borne'] = valid_data['Evse']
df['zone'] = valid_data['Zone']
df['jour'] = valid_data['date début de charge'].dt.day_name()
df['mois'] = valid_data['date début de charge'].dt.month_name()
```

#### Taux d'occupation par borne (%)

```python
borne_occup = df.groupby('borne')['period'].sum()*100/(365*24*60)
plt.bar(borne_occup.index, borne_occup)
plt.ylabel('Pourcentage du temps d\'occupation')
plt.title('Taux d\'occupation par borne')
plt.show()
```

#### Fréquentation par borne

```python
borne_occup = df.groupby('borne')['period'].count()
plt.bar(borne_occup.index, borne_occup)
plt.ylabel('nombre d\'occurence')
plt.title('Fréquentation par borne')
plt.show()
```
